```python
import pandas as pd
import numpy as np
```

```python
Tertiary = pd.read_csv("tertiary.csv")
```

```python
Tertiary.head()
Tertiary.columns
```

```python
from geopy.geocoders import Nominatim


geolocator = Nominatim()

Lat = []
Lon = []
n = len(Tertiary['City'])

Cities = Tertiary['City']
States = Tertiary['State']

for j in range(0, n):
    city = Cities[j]
    state = States[j]
    
    if city == "San Francisco":
        location = geolocator.geocode(94105)
        lat = location.latitude
        lon = location.longitude
        Lat.append(lat)
        Lon.append(lon)

    else:
        location = geolocator.geocode( city + " United States")
        lat = location.latitude
        lon = location.longitude
        Lat.append(lat)
        Lon.append(lon)

    
Tertiary['Lat'] = Lat
Tertiary['Lon'] = Lon
```

```python
Tertiary.head(10)
Tertiary[Tertiary.index == 0]
tertiary = Tertiary
```

```python
# Give the rank of the city for each factor

tertiary['Bar_Rank'] = tertiary['Bars'].rank(ascending = 0)
tertiary['Restaurant_Rank'] = tertiary['Restaurant'].rank(ascending = 0)
tertiary['Museums_Rank'] = tertiary['Museums'].rank(ascending = 0)
tertiary['Libraries_Rank'] = tertiary['Libraries'].rank(ascending = 0)
tertiary['Park_Rank'] = tertiary['Park_acres_per_1000_residents'].rank(ascending = 0)
tertiary['TopRes_Rank'] = tertiary['NumTop200Restau'].rank(ascending = 0)

tertiary
```

```python

# Set the rank of N/A to the (maximum ranking + 1) for this factor 
tertiary['Bar_Rank'].fillna(tertiary['Bar_Rank'].max()+1,inplace=True)
tertiary['Restaurant_Rank'].fillna(tertiary['Restaurant_Rank'].max()+1,inplace=True)
tertiary['Museums_Rank'].fillna(tertiary['Museums_Rank'].max()+1,inplace=True)
tertiary['Libraries_Rank'].fillna(tertiary['Libraries_Rank'].max()+1,inplace=True)
tertiary['Park_Rank'].fillna(tertiary['Park_Rank'].max()+1,inplace=True)
tertiary['TopRes_Rank'].fillna(tertiary['TopRes_Rank'].max()+1,inplace=True)
```

```python
# Calculat the sum of rankings for each city -- Total_Rank 
tertiary['Total_Rank'] = (tertiary['Bar_Rank']+tertiary['Restaurant_Rank']+tertiary['Museums_Rank']+tertiary['Libraries_Rank']+tertiary['Park_Rank']+tertiary['TopRes_Rank'])

# Rank the cities based on Total_Rank, the smallest number comes first -- Final_Rank
tertiary['Final_Rank'] = tertiary['Total_Rank'].rank(ascending = 1)

# Set the Final_Rank to be integers
tertiary['Final_Rank'] = tertiary['Final_Rank'].astype(int)

# Set reverse of Final_Rank to make plotting easier 
# (smaller ranking number will be projected to a bigger size)
tertiary['Reverse_Rank'] = 51 - tertiary['Final_Rank']
```

```python
tertiary = tertiary.sort_values('Final_Rank', ascending = 1)
tertiary.head(10)
```
